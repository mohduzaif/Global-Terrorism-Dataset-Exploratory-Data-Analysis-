#Data Wrangling.

```python
# Write your code to make the your dataset to analysis ready.
# Rename the columns.
global_terrorism_df.rename(columns={'eventid':'EventID','iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','success':'Success','suicide':'Suicide','provstate':'State','propvalue':'Propvalue','region_txt':'Region','attacktype1_txt':'AttackType','attacktype1':'Attack','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
global_terrorism_df['Killed'] = global_terrorism_df['Killed'].fillna(0)
global_terrorism_df['Wounded'] = global_terrorism_df['Wounded'].fillna(0)
```

```python
# Here we see the data is updated Successfully.
global_terrorism_df['Killed'].tail()
```

```python
# Now we see the data it will be updated Successfully.
global_terrorism_df['Wounded'].tail()
```

```python
# here we add the both killed and wounded people and update the killed column.
global_terrorism_df['Casualities'] = global_terrorism_df['Killed'] + global_terrorism_df['Wounded']
```

```python
# Here we see the data is updated Successfully.
global_terrorism_df['Casualities']
```

```python
# Indexing the columns.
global_terrorism_df = global_terrorism_df[['EventID','Year', 'Month', 'Day', 'Country', 'State', 'Region', 'city', 'Success', 'Suicide', 'Propvalue','AttackType','Attack', 'latitude', 'longitude', 'Wounded','Killed' ,'Target', 'Summary', 'Group', 'Target_type', 'Weapon_type', 'Motive', 'Casualities']]
```
